```python
import sys
sys.path.append('../src/')

from benchmark.tools import *
from benchmark.models import Detector
from benchmark.utils import TsvHandle, run_detection

import wider_dataset

# https://github.com/AyushExel/ml4face-detection/blob/master/Final-Face%20Detection%20Notebook.ipynb
import dlib
```

```python
train, test, validation = tfds.load('wider_dataset', split=['train', 'test', 'validation'])
```

```python
data = validation.take(30)
```

```python
class DlibDetector(Detector):
    def __init__(self):
        self.detector = dlib.get_frontal_face_detector()
        
    def detect(self, image):
        boxes = []
        confidence = []
        
        dets, scores, idx = self.detector.run(image, 1, -1)
        for i, d in enumerate(dets):
            if scores[i] > 0:
                x = d.left()
                y = d.top()
                x1 = d.right()
                y1 = d.bottom()
                
                boxes.append([x, y, x1, y1])
                confidence.append(scores[i])
        
        return boxes, confidence
    
dlib = DlibDetector()
```

```python
with TsvHandle('dlib') as tsv:
    run_detection(tsv, data, dlib, cv2.COLOR_RGB2GRAY, False, True, None)
```
